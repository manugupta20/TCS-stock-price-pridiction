```python
#Exploratry data analysis
plt.figure(figsize=(5,4))
plt.plot('close')
plt.title('TCS close price.',fontsize=15)
plt.ylabel('Price in doller')
plt.show()
```

```python
#Distribution plot
feature=['Open','High','Low','Close','Volume','Dividends','Stock Splits']
plt.subplots(figsize=(4,4))
for i, col in enumerate(feature):
    plt.subplot(2,4,i+1)
    sb.displot(df[col])
    plt.show()
```

```python
#pie chart
df['Open-Close'] = df['Open']-df['Close']
df['Low-High'] = df['Low']-df['High']
df['target'] = np.where(df['Close'].shift(-1)>df['Close'],1,0)
plt.pie(df['target'].value_counts().values, labels=[0,1],autopct='%1.1f%%')
plt.show()
```

```python
#heatmap of corelation
plt.figure(figsize=(10,10))
sb.heatmap(df.corr()>0.9,annot=True, cbar=False)
plt.show()
```

```python
x=['Open','Close']
y1=['High','Low']
y2=['Dividends','Volume']
plt.plot(x,y1, linestyle='dashed', marker='D', markersize='10', markerfacecolor='green', markeredgecolor='red')
plt.plot(x,y2, linestyle='dashed', marker='D', markersize='8', markerfacecolor='pink', markeredgecolor='blue')
plt.show()
```
